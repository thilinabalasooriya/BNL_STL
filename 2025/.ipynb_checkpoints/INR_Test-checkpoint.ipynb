```python
import torch
import torch.nn as nn
import math

class STAFSine(nn.Module):
    def __init__(self, coord_dim, hidden_dim, omega_0=30.0):
        """
        Args:
            coord_dim: dimensionality of input coords (e.g., 2 for (x, y))
            hidden_dim: number of neurons in the hidden layer
            omega_0: frequency factor for sine activation (as in SIREN)
        """
        super().__init__()
        self.transform = nn.Linear(coord_dim, hidden_dim)
        self.omega_0 = omega_0
        self.init_weights()

    def init_weights(self):
        # Initialize transform weights like SIREN recommends
        with torch.no_grad():
            self.transform.weight.uniform_(-1.0, 1.0)
            self.transform.bias.fill_(0.0)

    def forward(self, z, coords):
        """
        Args:
            z: (N, hidden_dim) pre-activation from Linear layer
            coords: (N, coord_dim) input coordinate vector
        Returns:
            (N, hidden_dim) STAF-Sine activated output
        """
        t = self.transform(coords)  # shape: (N, hidden_dim)
        return torch.sin(self.omega_0 * (z + t))
```

```python
class ImageINRModel(torch.nn.Module):

    def __init__(self):
        super(ImageINRModel, self).__init__()

        self.input = torch.nn.Linear(40, 256)
        self.l1 = torch.nn.Linear(256, 256)
        self.l2 = torch.nn.Linear(256, 256)
        self.l3 = torch.nn.Linear(256, 256)
        self.l4 = torch.nn.Linear(256, 256)
        self.output = torch.nn.Linear(256, 3)

        self.activation = torch.nn.ReLU()
        self.sig = torch.nn.Sigmoid()

    def forward(self, x):
        x = self.input(x)
        x = self.activation(x)
        x = self.l1(x)
        x = self.activation(x)
        x = self.l2(x)
        x = self.activation(x)
        x = self.l3(x)
        x = self.activation(x)
        x = self.l4(x)
        x = self.activation(x)
        x = self.output(x)
        x = self.sig(x)

        return x
```

```python
class ImageINRModelSTAF(torch.nn.Module):

    def __init__(self):
        super(ImageINRModelSTAF, self).__init__()

        self.input = torch.nn.Linear(40, 256)
        self.l1 = torch.nn.Linear(256, 256)
        self.l2 = torch.nn.Linear(256, 256)
        self.l3 = torch.nn.Linear(256, 256)
        self.l4 = torch.nn.Linear(256, 256)
        self.output = torch.nn.Linear(256, 3)
        self.activation = STAFSine(16, 256)

    def forward(self, x):
        x_init = x
        x = self.input(x)
        x = self.activation(x, x_init)
        x = self.l1(x)
        x = self.activation(x, x_init)
        x = self.l2(x)
        x = self.activation(x, x_init)
        x = self.l3(x)
        x = self.activation(x, x_init)
        x = self.l4(x)
        x = self.activation(x, x_init)
        x = self.output(x)

        return x
```

```python
class ImageINRDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        self.x = torch.tensor(dataframe[['x_pos', 'y_pos']].values, dtype=torch.float32)
        self.y = torch.tensor(dataframe[['R', 'G', 'B']].values, dtype=torch.float32)

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        return self.x[idx], self.y[idx]
```

```python
def fourier_encode(xy: torch.Tensor, freq: torch.Tensor) -> torch.Tensor:

    # Expand: (N, 2) → (N, 1, 2)
    xy = xy.unsqueeze(1)  # shape: (N, 1, 2)

    # (F,) → (1, F, 1), for broadcasting

    freq1 = freq[0:len(freq)//2 ]
    freq2 = freq[len(freq)//2:]
    
    print(freq1)
    print(freq2)
    
    freq1 = torch.tensor(freq1, dtype = torch.float32).view(1, -1, 1)
    freq2 = torch.tensor(freq2, dtype = torch.float32).view(1, -1, 1)

    # Multiply and encode
    scaled1 = 2 * torch.pi * (1/freq1) * xy  # shape: (N, F, 2)
    scaled2 = 2 * torch.pi * (1/freq2) * xy  # shape: (N, F, 2)

    sin_feat = torch.sin(scaled1)  # (N, F, 2)
    cos_feat = torch.cos(scaled2)  # (N, F, 2)

    # Concatenate and flatten last two dimensions: (N, F, 4) → (N, 4 * F)
    features = torch.cat([sin_feat, cos_feat], dim=-1)  # (N, F, 4)
    return features.view(xy.shape[0], -1)  # (N, 4 * F)
```

```python
from tqdm import tqdm
import pickle

epoch_losses = []
model = ImageINRModel().to(device)

criterion = torch.nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)


num_epochs = 2000

for epoch in tqdm(range(num_epochs)):
    model.train()

    epoch_loss = 0

    for i, (inputs, labels) in enumerate(train_dataloader):
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()

        outputs = model(inputs)
        loss = criterion(outputs, labels)
        losses.append(loss.item())
        

        epoch_loss += loss.item()

        loss.backward()
        optimizer.step()

 
    print(f"Epoch {epoch}, Loss: {epoch_loss / len(train_dataloader)}")
    
    if epoch % 20 == 0:
        epoch_losses.append(epoch_loss)
        pickle.dump(epoch_losses, open("losses.pkl", "wb"))
    
        checkpoint = {
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'loss': loss,
            # Add other relevant information as needed
        }

        torch.save(checkpoint, 'checkpoint.pth')
```
